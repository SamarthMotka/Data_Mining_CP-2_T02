# **Feature Engineering**

```python
# Categorical Features
categorical_columns = [feature for feature in df.columns if df[feature].dtypes == 'O']

# Numerical Features
numerical_columns = [feature for feature in df.columns if df[feature].dtypes != 'O']

# Create a table for categorical columns
categorical_table = tabulate({"Categorical columns": categorical_columns}, headers="keys", tablefmt="pretty")

# Create a table for numerical columns
numerical_table = tabulate({"Numerical columns": numerical_columns}, headers="keys", tablefmt="pretty")

# Print the tables
print("Categorical columns:")
print(categorical_table)

print("\nNumerical columns:")
print(numerical_table)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
```
